```python
%time metadata = pd.DataFrame([md for md in df.metadata])
metadata.head()
```

```python
%time abstract = pd.DataFrame([md for md in df.abstract])
abstract.head()
```

```python
%time body = pd.DataFrame([md for md in df.body_text])
body.head()
```

```python
df1 = df[['paper_id','discipline','metadata','abstract','body_text']]
df1.head()
```

```python
df1 = pd.concat([df1, metadata], axis=1)
```

```python
df1.head()
```

```python
df1 = df1.drop(['id', 'metadata'], axis=1)
```

```python
df1 = pd.concat([df1, abstract], axis=1)
df1.head()
```

```python
df1 = pd.concat([df1, body], axis=1)
```

```python
df2 = df1[['paper_id','discipline','submitter','authors','title','comments','journal-ref','doi','categories','abstract','update_date','versions','authors_parsed','section','text']]
df2.head()
```

```python
csv_file_path = 'dataset.csv'

# Convert the DataFrame to a CSV file
df2.to_csv(csv_file_path, index=False)
```
